```python
from Bio import SeqIO, Seq
for dataset_name, data_dict in total_datasets.items():
    save_path = save_data_path / dataset_name
    save_path.mkdir(parents=True, exist_ok=True)
    fasta_path = save_path / "{}.fasta".format(dataset_name)
    SeqIO.write(SeqIO.SeqRecord(Seq.Seq(data_dict["wildtype"]), dataset_name), fasta_path.open("w"), "fasta")
    mut_fitness_path = save_path / "{}.csv".format(dataset_name)
    data_dict["mut_fitness"].to_csv(mut_fitness_path.open("w"), index=False)
```
